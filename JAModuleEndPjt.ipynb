```python
age=a1['Age'].value_counts().head(30)
age
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Age",data=a1)
plt.title("Age Group Category")
plt.show()
```

```python
bins= [15,20,25,30,35,40]
labels=["15-20","21-25","26-30","31-35","36-40"]
a1["Age_Group"]=pd.cut(a1["Age"],bins=bins,labels=labels)
sns.countplot(x="Age_Group",data=a1)
plt.title("Age Group Category")
plt.show()
```

```python

#4. Find out under which team and position, spending in terms of salary is high.
```

```python
Team_salary=a1.groupby(["Team","Position"])["Salary"].sum()
Team_salary
```

```python
Team_salary.sort_values().tail()
```

```python
highest=Team_salary.max()
highest
highest_spending=Team_salary.loc[Team_salary==highest]
print(f"Team and position, spending in terms of salary is high :{highest_spending}")
```

```python

#5.Find if there is any correlation between age and salary , represent it visually.
```

```python
sns.scatterplot(x="Age",y="Salary",data=a1)
plt.xlabel("Age",color='r')
plt.ylabel("Salary",color='r')
plt.xticks([18,20,22,24,26,28,30,32,34,36,38,40])
plt.title("Correlation Between Age & Salary",color='r')
plt.show()
```
